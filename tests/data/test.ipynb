```python
from tsmaccountingmanager.backend.models import Sale, Purchase, extract_item_id, process_price

# a function to process the sales and expenses dataframes and print the code that can be used to create the objects
def process_and_print(sales, expenses):
    # create a list of sales objects
    sales_list = []
    for index, row in sales.iterrows():
        itemId = extract_item_id(row["itemString"])
        stackSize = row["stackSize"]
        quantity = row["quantity"]
        price, total = process_price(row["price"], stackSize)
        time = row["time"]
        source = row["source"]
        sales_list.append(Sale(
            item=itemId,
            stackSize=stackSize,
            quantity=quantity,
            price=price,
            total=total,
            time=time,
            source=source,
        ))



    # create a list of expenses objects
    expenses_list = []
    for index, row in expenses.iterrows():
        itemId = extract_item_id(row["itemString"])
        stackSize = row["stackSize"]
        quantity = row["quantity"]
        price, total = process_price(row["price"], stackSize)
        time = row["time"]
        source = row["source"]
        expenses_list.append(Purchase(
            item=itemId,
            stackSize=stackSize,
            quantity=quantity,
            price=price,
            total=total,
            time=time,
            source=source,
        ))

    # print the code that would create the sales_list

    print("sales_list = [")
    for sale in sales_list:
        print(f"    Sale(item={sale.item}, stackSize={sale.stackSize}, quantity={sale.quantity}, price={sale.price}, total={sale.total}, time=process_timestamp({sale.time}), source=Source.{sale.source.upper()}),")
    print("]")

    # print the code that would create the expenses_list
    print("expenses_list = [")
    for expense in expenses_list:
        print(f"    Purchase(item={expense.item}, stackSize={expense.stackSize}, quantity={expense.quantity}, price={expense.price}, total={expense.total}, time=process_timestamp({expense.time}), source=Source.{expense.source.upper()}),")
    print("]")
```

```python
process_and_print(sales, expenses)
```

```python
from tsmaccountingmanager.backend.models import process_timestamp, Source
from datetime import datetime

purchases_list = [
    Purchase(
        item=179314,
        stackSize=498,
        quantity=498,
        price=14.5683,
        total=7255.013400000001,
        time=process_timestamp(1641925078),
        source=Source.AUCTION,
    ),
    Purchase(
        item=173033,
        stackSize=993,
        quantity=993,
        price=14.1309,
        total=14031.9837,
        time=process_timestamp(1641925080),
        source=Source.AUCTION,
    ),
    Purchase(
        item=53010,
        stackSize=950,
        quantity=950,
        price=1.1372,
        total=1080.34,
        time=process_timestamp(1642191158),
        source=Source.AUCTION,
    ),
]

for i in range(0, len(purchases_list)):
    print(f"{i} : {purchases_list[i].time}")

# print them in order
for purchase in sorted(purchases_list, key=lambda purchase: purchase.time):
    print(purchase.time)

start = process_timestamp(purchases_list[0].time.timestamp()-1)
end = process_timestamp(purchases_list[1].time.timestamp()+1)

print(end < purchases_list[2].time)
print(start < purchases_list[0].time)
print(purchases_list[0].time < purchases_list[1].time)
print(purchases_list[1].time < end)

start_2 = process_timestamp(purchases_list[2].time.timestamp()-1)
end_2 = process_timestamp(purchases_list[2].time.timestamp()+1)

print(start_2 < purchases_list[2].time)
print(purchases_list[2].time < end_2)
print(start_2 > purchases_list[1].time)

# print the code that would create these two pairs of timestamps
print(f"dates = [(process_timestamp({start.timestamp()}), process_timestamp({end.timestamp()})), (process_timestamp({start_2.timestamp()}), process_timestamp({end_2.timestamp()}))]")
```
